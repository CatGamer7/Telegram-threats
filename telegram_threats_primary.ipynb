# Emo-NER 

```python
import numpy as np
from tqdm import tqdm
from json import dumps, loads, JSONEncoder


tqdm.pandas()
```

```python
from transformers import (
    AutoTokenizer, 
    AutoModelForTokenClassification,
    AutoModelForSequenceClassification, 
    pipeline
)

%env HF_HUB_DISABLE_SYMLINKS_WARNING=1
```

```python
import torch
print(torch.cuda.is_available())
```

```python
ner_name = "Babelscape/wikineural-multilingual-ner"
ner_tokenizer = AutoTokenizer.from_pretrained(ner_name)
ner_model = AutoModelForTokenClassification.from_pretrained(ner_name).to("cuda")
ner_pipe = pipeline("ner", model=ner_model, tokenizer=ner_tokenizer, grouped_entities=True, device=0)
```

```python
df_l_clean["named_entities"] = df_l_clean["reply"].progress_apply(
    lambda x: ner_pipe(x) if isinstance(x, str) else []
)
```

```python
class NumpyEncoder(JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.floating):
            return float(obj)
        else:
            return super(NumpyEncoder, self).default(obj)
```

```python
df_l_clean["named_entities_json"] = df_l_clean["named_entities"].apply(
    lambda x: dumps(x, cls=NumpyEncoder)
)
```

```python
emotion_name = "j-hartmann/emotion-english-distilroberta-base"
emotion_tokenizer = AutoTokenizer.from_pretrained(emotion_name)
emotion_model = AutoModelForSequenceClassification.from_pretrained(emotion_name).to("cuda")
emotion_pipe = pipeline("text-classification", model=emotion_model, tokenizer=emotion_tokenizer, top_k=None, device=0)
```

```python
def sample(source_text, model, sampling_size = 500):

    # If 4-byte UTF-8 characters are present - make smaller steps to fit text into tokenizer window
    if len(source_text[:sampling_size].encode()) > 500:
        sampling_size = min(sampling_size // 4, 100)

    try:
        return model(source_text[:sampling_size])[0]
    except RuntimeError:
        return None
```

```python
df_l_clean["emotion"] = df_l_clean["reply"].progress_apply(
    lambda x: sample(x, emotion_pipe)
)
```

```python
df_l_clean["emotion_json"] = df_l_clean["emotion"].apply(
    lambda x: dumps(x, cls=NumpyEncoder)
)
```

```python
df_l_clean.to_csv("labeled_ner_emo.csv", index=False)
```
